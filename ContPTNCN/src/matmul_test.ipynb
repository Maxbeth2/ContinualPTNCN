```python
import tensorflow as tf
import numpy as np

```

```python
x = tf.constant([[1.,0.,1.,0.]])

smth = tf.constant([[-1., 0.]])

corr = tf.constant([[1., 1.]])

w = tf.matmul(x, smth, transpose_a=True)

pred = tf.matmul(x, w)

err = tf.subtract(corr, pred)
print(corr)
print(pred)
print(err)

for _ in range(20):

    dW = tf.matmul(x, err,  transpose_a=True) * 0.1
    w = tf.add(w, dW)
    print()

    pred = tf.matmul(x, w)
    err = tf.subtract(corr, pred)
    print(corr)
    print(pred)
    print(err)
```
